1) Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

```python
dados_desafio.isnull().sum()
```

```python
dados_desafio['Notas'].mean()
```

```python
nulos = dados_desafio.query('Notas.isnull()').index
nulos
```

```python
dados_sem_nulos = dados_desafio.drop(nulos, axis=0)
dados_sem_nulos['Notas'].mean()
```

```python
dados_desafio = dados_desafio.fillna(dados_sem_nulos['Notas'].mean())
dados_desafio
```

2) Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

```python
alice = dados_desafio['Nome'] != 'Alice'
carlos = dados_desafio['Nome'] != 'Carlos'
filtro_nome = (alice) & (carlos)
filtro_nome
```

```python
dados_eliminar = dados_desafio.query('Nome == "Alice" | Nome == "Carlos"').index
dados_desafio.drop(dados_eliminar, axis=0, inplace=True)
dados_desafio
```

```python
dados_desafio = dados_desafio[filtro_nome]
dados_desafio
```

3) Aplique um filtro que selecione apenas os alunos que foram aprovados.

```python
dados_desafio.query('Notas >= 7')
```

```python
alunos_aprovados = dados_desafio.query('Notas >= 7')
```

4) Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

Extra: Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade, tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0.

```python
alunos_aprovados = alunos_aprovados.replace({'Notas': 7.0}, 8.0)
alunos_aprovados
```

1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
dados_desafio['Pontos_extras'] = dados_desafio['Notas'].apply(lambda x: x*0.4)
dados_desafio.head()
```

2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

```python
dados_desafio['Notas_finais'] = dados_desafio['Pontos_extras'] + dados_desafio['Notas']
dados_desafio.head()
```

3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

- True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
- False: caso o aluno esteja reprovado (nota final deve ser menor que 6).

```python
dados_desafio['Aprovado_final'] = dados_desafio['Notas_finais'].apply(lambda x: True if x >= 6 else False)
dados_desafio.head()
```

4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
filtro_1 = (dados_desafio['Aprovado'] == False) & (dados_desafio['Aprovado_final'] == True)
dados_desafio[filtro_1]
```
